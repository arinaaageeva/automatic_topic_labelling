```python
import sys
import shutil
import pickle as pkl

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

from tqdm import tqdm
from sklearn.metrics import accuracy_score, f1_score

sys.path.append('../src')
from topic_modeling import *
```

```python
def visualisation(level_names, score_names, model_scores, ylims=None, figsize=(10, 10)):
    
    num_levels = len(level_names)
    num_scores = len(score_names)
    
    fig, axis = plt.subplots(num_scores, num_levels, sharey=False, figsize=figsize)

    for model_name, scores in model_scores.items():
        for i, score in enumerate(scores.values()):
            for j, level_score in enumerate(score):
                axis[i][j].plot(level_score, label=model_name)
    
    for i, score_name in enumerate(score_names):
        for j, level_name in enumerate(level_names):
            axis[i][j].set_title(f'{score_name} ({level_name})')
            axis[i][j].legend()
    
    plt.show()
```

```python
dictionary = artm.Dictionary(dictionary_path='../data/interim/X_dictionary.dict')

with open(f'../data/interim/X', 'rb') as fl:
    X = pkl.load(fl)

with open(f'../data/interim/y_level_0', 'rb') as fl:
    y_level_0 = pkl.load(fl)
    
topic_names_level_0 = y_level_0.unique().tolist()
    
with open(f'../data/interim/y_level_1', 'rb') as fl:
    y_level_1 = pkl.load(fl)
    
topic_names_level_1 = y_level_1.unique().tolist()

y = pd.concat([y_level_0, y_level_1], axis=1)

level_names = y.columns

level_topic_names = dict(zip(level_names, [topic_names_level_0, topic_names_level_1]))

class_ids=['title_pers', 'title_orgs', 'title_locs', 'title_tokens', 'text_pers', 'text_orgs', 'text_locs', 'text_tokens']
default_class_ids = dict(zip(class_ids, len(class_ids)*[1.0]))

kfold = KFoldTM(random_state=42)
```

```python
def f1_weighted_score(x, y, ignore_background=True):
    
    if ignore_background:

        background_flags = x.apply(lambda x: 'background' in x)

        x = x[~background_flags]
        y = y[~background_flags] 
    
    return f1_score(x, y, average='weighted')

background_score = lambda x, y: x.apply(lambda x: 'background' in x).sum()
```

### Isolated levels vs. Hierarchy

```python
model_scores = dict()

model_scores['only level 0'] = kfold.validation(X, 
                                                y.drop('hr_level_1', axis=1), 
                                                {'accuracy':accuracy_score, 'f1_score':f1_weighted_score}, 
                                                class_ids=default_class_ids, 
                                                level_names=[level_names[0]], 
                                                topic_names=[topic_names_level_0])

model_scores['only level 1'] = kfold.validation(X, 
                                                y.drop('hr_level_0', axis=1), 
                                                {'accuracy':accuracy_score, 'f1_score':lambda x, y: f1_score(x, y, average='weighted')}, 
                                                class_ids=default_class_ids, 
                                                level_names=[level_names[1]], 
                                                topic_names=[topic_names_level_1])

model_scores = {'isolated levels': {name: np.concatenate([model_scores['only level 0'][name], model_scores['only level 1'][name]]) for name in ['accuracy', 'f1_score']}}
```

```python
model_scores['hierarchy'] = kfold.validation(X, 
                                             y, 
                                             {'accuracy':accuracy_score, 'f1_score':lambda x, y: f1_score(x, y, average='weighted')}, 
                                             class_ids=default_class_ids, 
                                             level_names=level_names, 
                                             topic_names=[topic_names_level_0, topic_names_level_1])
```

```python
visualisation(level_names, ['Accuracy', 'F1-score'], model_scores, figsize=(20, 10))
```

### Fit Psi

```python
with open('../data/interim/X.txt', 'w') as fl:
    fl.write('\n'.join(X.index + ' ' + X.values)+'\n')
    
X_batches = artm.BatchVectorizer(data_path='../data/interim/X.txt', data_format='vowpal_wabbit', target_folder='../data/interim/X_batches')
```

```python
hierarchy = TopicModel(dictionary, 
                       default_class_ids, 
                       level_names=level_names, 
                       topic_names=[topic_names_level_0, topic_names_level_1], 
                       num_collection_passes=10)

hierarchy.fit(X_batches, y)
```

```python
hierarchy.model[1].get_psi()
```

```python
parent_topics_proportion = {}

for topic_name, parent_topics in y.groupby('hr_level_1')['hr_level_0'].unique().iteritems():
    parent_topic_proportion = np.zeros(len(topic_names_level_0))
    parent_topic_proportion[[topic_names_level_0.index(parent_topic_name) for parent_topic_name in parent_topics]] = 1
    
    parent_topics_proportion[topic_name] = parent_topic_proportion.tolist()
```

```python
parent_topics_proportion
```

```python
hierarchy = TopicModel(dictionary, 
                       default_class_ids, 
                       level_names=level_names, 
                       topic_names=[topic_names_level_0, topic_names_level_1], 
                       spars_psi_tau=1.0, 
                       parent_topics_proportion=[parent_topics_proportion], 
                       num_collection_passes=10)

hierarchy.fit(X_batches, y)
```

```python
hierarchy.model[1].get_psi()
```

```python
shutil.rmtree('../data/interim/X_batches')
```

```python
model_scores = dict()
for spars_psi_tau in [0, 1.0]:
    
    name = f'sparse phi {spars_psi_tau}'
    model_scores[name] = kfold.validation(X, 
                                          y, 
                                          {'accuracy':accuracy_score, 'f1_score':lambda x, y: f1_score(x, y, average='weighted')}, 
                                          class_ids=default_class_ids,
                                          level_names=level_names, 
                                          topic_names=[topic_names_level_0, topic_names_level_1],
                                          parent_topics_proportion=[parent_topics_proportion],
                                          spars_psi_tau=spars_psi_tau)
```

```python
visualisation(level_names, ['Accuracy', 'F1-score'], model_scores, figsize=(20, 10))
```

### Background, Smooth and Decorrelation

```python
model_scores = dict()
for smooth_background_tau in [0, 0.3, 0.5, 0.7, 0.9, 1]:
    
    name = f'smooth back {smooth_background_tau}'
    model_scores[name] = kfold.validation(X, 
                                          y, 
                                          {'background_score':background_score, 'f1_score':f1_weighted_score}, 
                                          class_ids=default_class_ids,
                                          level_names=level_names, 
                                          topic_names=[topic_names_level_0, topic_names_level_1],
                                          num_backgrounds=[1, 1], 
                                          smooth_background_tau=smooth_background_tau,
                                          spars_psi_tau=1.0,
                                          parent_topics_proportion=[parent_topics_proportion])
```

```python
visualisation(level_names, ['Background', 'F1-score'], model_scores, figsize=(20, 10))
```

```python
model_scores = dict()
for decorrelate_phi_tau in [0, 0.05, 0.1, 0.15, 0.3]:
    
    name = f'decorrelate {decorrelate_phi_tau}'
    model_scores[name] = kfold.validation(X, 
                                          y, 
                                          {'background_score':background_score, 'f1_score':f1_weighted_score}, 
                                          class_ids=default_class_ids,
                                          level_names=level_names, 
                                          topic_names=[topic_names_level_0, topic_names_level_1],
                                          num_backgrounds=[1, 1], 
                                          smooth_background_tau=1,
                                          decorrelate_phi_tau=decorrelate_phi_tau,
                                          spars_psi_tau=1.0,
                                          parent_topics_proportion=[parent_topics_proportion])
```

```python
visualisation(level_names, ['Background', 'F1-score'], model_scores, figsize=(20, 10))
```

```python
for index, level_tokens in enumerate(hierarchy.get_top_tokens()):
    print(f'LEVEL {index}\n')
    for topic_name, modalities in level_tokens.items():
        print(topic_name.upper())
        for modality, tokens in modalities.items():
            tokens = '; '.join(tokens)
            print(f'{modality}: {tokens}')
        print()
```
