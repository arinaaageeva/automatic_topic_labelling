```python
import sys
import artm
import itertools

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

from tqdm import tqdm
from collections import Counter
from sklearn.metrics import *

sys.path.append('../src')
from topic_modeling import *
```

```python
dictionary = artm.Dictionary()
dictionary.gather(data_path='../data/interim/batches')
batch_vectorizer = artm.BatchVectorizer(data_path='../data/interim/batches', data_format='batches')

dictionary_split_ner = artm.Dictionary()
dictionary_split_ner.gather(data_path='../data/interim/batches_split_ner')
batch_vectorizer_split_ner = artm.BatchVectorizer(data_path='../data/interim/batches_split_ner', data_format='batches')

class_ids=['title_pers', 'title_orgs', 'title_locs', 'title_tokens', 'text_pers', 'text_orgs', 'text_locs', 'text_tokens']
```

**Построение первого уровня иерархии**

```python
articles = pd.read_csv('../data/interim/articles_new.csv')
labels_true = articles[['id', 'hr_level_0']]

articles_ids = labels_true.id.apply(str).values
labels_true = labels_true.hr_level_0.values

num_topics = len(np.unique(labels_true))
print(f'Первый уровень иерархии содержит {num_topics} тем')
```

Класстеризуем статьи без регуляризаторов и смотрим на метрики: 

    * Adjusted Rand index
    * Adjusted Mutual Information
    * Homogeneity score 
    * Completeness score

```python
#костыль
def adjusted_mutual_info_score_(x, y):
    return adjusted_mutual_info_score(x, y, average_method='arithmetic')

metrics = {'adjusted rand score': adjusted_rand_score,
           'adjusted mutual info score': adjusted_mutual_info_score_,
           'homogeneity score': homogeneity_score,
           'completeness score': completeness_score}

def scores_visualisation(scores):
    
    n_metrics = len(scores)    
    fig, axes = plt.subplots(n_metrics // 2 + n_metrics % 2, 2, figsize=(20, 10))
    
    for index, (metric_name, models_scores) in enumerate(scores.items()):
        
        index_i, index_j = index // 2, index % 2
        for model_name,  model_scores in models_scores.items():
            axes[index_i, index_j].plot(model_scores, label=model_name)
            
        axes[index_i, index_j].set_ylabel(metric_name)    
        axes[index_i, index_j].legend()

    fig.show() 
```

Исследуем какая из модальностей даёт лучшее качество

```python
class_ids_weights = np.eye(len(class_ids))
currents_class_ids = [dict(zip(class_ids, weights)) for weights in class_ids_weights]

scores = {}
for current_class_ids in tqdm(currents_class_ids):
    
    model_name = '; '.join([f'dec {class_id}' + ('' if tau == 1 else f':{tau:.2f}') 
                            for class_id, tau in decorrelat_phi_class_ids.items() if tau > 0])
    
    topic_model = TopicModeling(dictionary_split_ner, num_topics=num_topics, num_collection_passes=1, class_ids=current_class_ids)
    
    for _ in range(50):

        topic_model.fit(batch_vectorizer_split_ner)
        labels_pred = topic_model.labels_[articles_ids].values
        
        for metric_name, metric in metrics.items():
            scores.setdefault(metric_name, {})
            scores[metric_name].setdefault(model_name, [])
            scores[metric_name][model_name].append(metric(labels_true, labels_pred))
            
scores_visualisation(scores)
```

Видим, что лучшее качество получено на токенах из текста, далее подберем необходимые параметры регуляризации

```python
current_class_ids = dict(zip(class_ids, [0, 0, 0, 0, 0, 0, 0, 1]))
decorrelats_phi_class_ids = [dict(zip(class_ids, tau*np.array([0, 0, 0, 0, 0, 0, 0, 1]))) for tau in [0, 0.01, 0.05, 0.1]]

scores = {}
for decorrelat_phi_class_ids in tqdm(decorrelats_phi_class_ids):
    
    model_name = '; '.join([f'smooth {class_id}: {tau:.2f}' for class_id, tau in decorrelat_phi_class_ids.items() if tau > 0])
    
    topic_model = TopicModeling(dictionary_split_ner, num_topics=num_topics, num_collection_passes=1, 
                                class_ids=current_class_ids, decorrelat_phi_class_ids=decorrelat_phi_class_ids)
    
    for _ in range(50):

        topic_model.fit(batch_vectorizer_split_ner)
        labels_pred = topic_model.labels_[articles_ids].values
        
        for metric_name, metric in metrics.items():
            scores.setdefault(metric_name, {})
            scores[metric_name].setdefault(model_name, [])
            scores[metric_name][model_name].append(metric(labels_true, labels_pred))
        
scores_visualisation(scores)
```

```python
current_class_ids = dict(zip(class_ids, [0, 0, 0, 0, 0, 0, 0, 1]))
decorrelat_phi_class_ids = dict(zip(class_ids, [0, 0, 0, 0, 0, 0, 0, 0.01]))
smooths_background_class_ids = [dict(zip(class_ids, tau*np.array([0, 0, 0, 0, 0, 0, 0, 1]))) for tau in [0, 0.1, 0.3, 0.05, 0.7, 0.9, 1]]

scores = {}
for smooth_background_class_ids in tqdm(smooths_background_class_ids):
    
    model_name = '; '.join([f'smooth {class_id}: {tau:.2f}' for class_id, tau in smooth_background_class_ids.items() if tau > 0])
    
    topic_model = TopicModeling(dictionary_split_ner, num_topics=num_topics, num_background_topics=1, num_collection_passes=1, 
                                class_ids=current_class_ids, 
                                decorrelat_phi_class_ids=decorrelat_phi_class_ids, 
                                smooth_background_class_ids=smooth_background_class_ids)
    
    for _ in range(50):

        topic_model.fit(batch_vectorizer_split_ner)
        
        labels_pred = topic_model.labels_[articles_ids].values
        background_pred = np.array(['background' in label for label in labels_pred])
        
        for metric_name, metric in metrics.items():
            scores.setdefault(metric_name, {})
            scores[metric_name].setdefault(model_name, [])
            scores[metric_name][model_name].append(metric(labels_true[~background_pred], labels_pred[~background_pred]))
            
        scores.setdefault('background size', {})
        scores['background size'].setdefault(model_name, [])
        scores['background size'][model_name].append(sum(background_pred))
            
scores_visualisation(scores)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder().fit(labels_true)
labels_true_encode = label_encoder.transform(labels_true)

num_background_topics = 1

doc_topic_coef = np.zeros((len(labels_true_encode), num_topics+num_background_topics))
doc_topic_coef[range(len(labels_true_encode)), labels_true_encode] = 1.0

topic_names = [f'topic_{index}' for index in range(num_topics)]
background_names = [f'background_{index}' for index in range(num_background_topics)]
        
model = artm.ARTM(dictionary=dictionary, topic_names=topic_names+background_names, theta_columns_naming='title', cache_theta=True)

#model.scores.add(artm.TopTokensScore(name='text_tokens_top_tokens', class_id='text_tokens', num_tokens=10))
model.regularizers.add(artm.SmoothSparseThetaRegularizer(name='SST', tau=100.0, doc_titles=articles_ids, doc_topic_coef=doc_topic_coef.tolist()))

model.fit_offline(batch_vectorizer=batch_vectorizer, num_collection_passes=10)
```
